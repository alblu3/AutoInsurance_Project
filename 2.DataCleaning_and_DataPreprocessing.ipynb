## Import Libraries

```python
# Import basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

## Load Dataset

```python
# Load Dataset
df = pd.read_csv('AutoInsurance.csv')
df.head()
```

## Checking Dataset  
 
*   Is there any missing value
*   Column only has 1 value 
*   Irrelevant column (not needed for analysis)




```python
# Create a dataframe to extract the wanted information
pd.DataFrame({'Feature': df.columns, 'DataType': df.dtypes,
             'Total Null Value': df.isna().sum(), 'Unique': df.nunique(),
             'Unique Value Sample': [list(df[i].drop_duplicates().sample()) for i in df.columns]}).reset_index(drop=True)
```

**Summary:**
1. No missing value in this dataset 
2. No column that only has 1 value and all columns/features will be used for feature selection
